# Visualizing line segments in $C[a,b]$ #

Visualizing the "line segment" between two points in the unit ball of continuous function space $C[a,b]$. 
Kreyszig problem 2.2.11 shows that the unit ball (space of points with norm $\leq 1$) in any normed space is convex, that is, the line segment between any two points in the unit ball is also in the unit ball.  
This is intuitively easy to see for $\textbf{R}^2$ but what about a more exotic space like $C[a,b]$ with norm max$|f(t)|$?
To illustrate with an example, let's use $[a,b] = [-1,1]$ and $f(t) = cos(2t)$, and $g(t) = t^3$. 
So the line segment will be all functions (1):
$$\alpha \text{cos}(2t) + (1-\alpha) t^3$$ 
evaluated from $-1 \leq t \leq 1$ for $0 \leq \alpha \leq 1$. 
Both functions are bounded by one so they are in the unit ball by definition, and from problem 2.2.11, we know that all points (functions) in the line segment will be in the unit ball, i.e. bounded by 1. 
Let's watch this happen with an animation. 

```python
#import libraries
import random
import math

import numpy as np
import matplotlib.pyplot as plt


import itertools
import matplotlib.animation as animation

from IPython.display import HTML
```

Plot the starting functions:

```python
%matplotlib inline

t = np.linspace(-1,1,40)

f = np.cos(t*np.pi)
g = t**3

plt.plot(t, f)
plt.plot(t, g)

plt.show()
```

Now add the functions as in (1) and sweep through $\alpha$ : 

```python
%matplotlib agg


#fig, ax = plt.subplots(1, 1, figsize = (6, 6))

t = np.linspace(-1,1,40)

f = np.cos(t*np.pi)
g = t**3


fig, ax = plt.subplots(1, 1, figsize = (6, 6))
T=21

ax.set_xlim([0,1]) # fix the x axis
ax.set_ylim([-1.2,1.2]) # fix the y axis

# plot the rest
def animate(i):
    ax.cla() # clear the previous image
    plt.title(r"$\alpha$ cos$(2t) + (1-\alpha)$ $t^3$")
    ax.set_xlim([-1.1,1.1]) # fix the x axis
    ax.set_ylim([-1.2,1.2]) # fix the y axis
    alpha = i/(T-1)
    ax.plot(t,alpha*f + (1-alpha)*g)
    plt.annotate(r"$\alpha=$"+str(alpha), xy=(0.05, 0.95), xycoords='axes fraction')


anim = animation.FuncAnimation(fig, animate, frames = T, interval = 250, blit = False)
HTML(anim.to_jshtml())
```

The combination morphs from the cubic to the cosine function, while keeping its bound of 1.
